```python
# Import the libraries
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
df = iris.iloc[:,:4]
df
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=12)
dbscan.fit(df)
```

```python
dbscan.labels_
```

```python
# -1 are the noise points in our dataset and the rest are the clusters
```

```python
cl = pd.DataFrame(dbscan.labels_, columns=['cluster'])
```

```python
cluster = pd.concat([iris,cl],axis=1)
```

```python
cluster.groupby('cluster').agg(['mean'])
```

```python
cluster.groupby('species').agg(['mean'])
```
